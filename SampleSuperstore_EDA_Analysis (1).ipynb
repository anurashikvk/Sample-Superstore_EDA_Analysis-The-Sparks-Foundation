# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Selecting only relevant columns
relevant_columns = df[['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit']]

# Encoding categorical columns
encoder = LabelEncoder()
encoded_df = relevant_columns.apply(encoder.fit_transform)

# Compute correlation matrix
correlation_matrix = encoded_df.corr()

# Create correlation plot
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})

# Add title
plt.title('Correlation Plot')

# Show plot
plt.show()
```

### Correlation Analysis:

In the correlation heatmap above, we observe the following correlations between 'Profit' and other variables:

1. **Sales (0.50)**:
   - There is a moderate positive correlation between 'Profit' and 'Sales' (0.50).
   - This suggests that higher sales volume tends to contribute positively to profitability.

2. **Quantity (0.23)**:
   - There is a weak positive correlation between 'Profit' and 'Quantity' (0.23).
   - This indicates that there is some association between the quantity of products sold and the profit earned, but it's not as strong as the correlation with sales.

3. **Sub-Category (0.03)**:
   - There is a very weak positive correlation between 'Profit' and 'Sub-Category' (0.03).
   - This suggests a minimal association between the type of product being sold and profitability.

4. **Category (0.18)**:
   - There is a weak positive correlation between 'Profit' and 'Category' (0.18).
   - This indicates that the broad category of products being sold has some influence on profitability, but the correlation is not very strong.

5. **Region (0.15)**:
   - There is a weak positive correlation between 'Profit' and 'Region' (0.15).
   - This suggests that the region where sales are made has a slight influence on profitability.

6. **Segment (0.02)**:
   - There is a very weak positive correlation between 'Profit' and 'Segment' (0.02).
   - This indicates that the market segment has minimal impact on profitability.

7. **Shipment Mode (-0.01)**:
   - There is a very weak negative correlation between 'Profit' and 'Ship Mode' (-0.01).
   - This suggests that the shipment mode has almost no impact on profitability.

8. **Discount (-0.57)**:
   - There is a moderate negative correlation between 'Profit' and 'Discount' (-0.57).
   - This indicates that as the discount given on products increases, profitability tends to decrease. It suggests that offering higher discounts negatively impacts profit margins.

In summary, the analysis of the correlations suggests that 'Sales' and 'Discount' have the most significant influence on profitability, with 'Quantity' and 'Category' also showing some association, albeit weaker. Other factors such as 'Sub-Category', 'Region', 'Segment', and 'Shipment Mode' have minimal impact on profitability based on the provided correlation coefficients.


```python
# Set up the figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 5))

# Plot sales distribution
sns.barplot(x='Category', y='Sales', data=df, ax=axs[0], color='blue', alpha=0.5)
axs[0].set_title('Sales Distribution across Categories')
axs[0].set_ylabel('Sales')
axs[0].set_xlabel('Category')
axs[0].tick_params(axis='x', rotation=45)

# Plot Discount distribution
sns.barplot(x='Category', y='Discount', data=df, ax=axs[1], color='red', alpha=0.5)
axs[1].set_title('Discount Distribution across Categories')
axs[1].set_ylabel('Discount')
axs[1].set_xlabel('Category')
axs[1].tick_params(axis='x', rotation=45)

# Plot profit distribution
sns.barplot(x='Category', y='Profit', data=df, ax=axs[2], color='green', alpha=0.5)
axs[2].set_title('Profit Distribution across Categories')
axs[2].set_ylabel('Profit')
axs[2].set_xlabel('Category')
axs[2].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculate profit margin
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100

# Calculate mean profit margin for each category
profit_margin_by_category = df.groupby('Category')['Profit Margin'].mean().reset_index()

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plot pie chart for profit margin distribution
ax.pie(profit_margin_by_category['Profit Margin'], labels=profit_margin_by_category['Category'], autopct='%1.1f%%', startangle=140)

# Add title
ax.set_title('Profit Margin Distribution across Categories')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Show plot
plt.show()
```
